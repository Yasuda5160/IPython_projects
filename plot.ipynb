```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(-10, 10)

plt.plot(x, np.sin(x), "o-", color = "red", alpha = 0.4)
plt.show()
```

```python
rg = np.random.default_rng()
x = rg.normal(0, 1, 5000)

plt.hist(x, bins = 20)
plt.show()
```

```python
rg = np.random.default_rng()
xy = rg.standard_normal(size = (100, 2))

plt.scatter(xy[:, 0], xy[:, 1], marker = "x", alpha = 0.8)
```

```python
x_image = rg.uniform(size = (32, 32))

plt.imshow(x_image, cmap = "gray")
plt.colorbar()#colorbar is available
plt.show
```

```python
x_image = rg.uniform(size = (32, 32, 3))

plt.imshow(x_image)
plt.show()
```

```python
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
u, v = np.meshgrid(x, y)

plt.quiver(x, y, u, v, color = "blue")
```

```python
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contour(X, Y, Z)
plt.colorbar
```

```python
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contourf(X, Y, Z)
plt.colorbar
```

```python
x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.quiver(x, y, X, Y, color = "blue")
plt.contour(X, Y, Z)
plt.colorbar

plt.colorbar()
plt.show()
```

```python
x = np.linspace(.1, 10)

plt.figure(figsize=(14,4))

plt.subplot(2,3,1)
plt.plot(x, np.sin(x))

plt.subplot(2,3,2)
plt.plot(x, np.cos(x))

plt.subplot(2,3,3)
plt.plot(x, np.tan(x))

plt.subplot(2,3,4)
plt.plot(x, np.exp(x))

plt.subplot(2,3,5)
plt.plot(x, np.log(x))

plt.subplot(2,3,6)
plt.plot(x, np.tanh(x))
```

```python
x = np.linspace(.1, 10)

fig = plt.figure(figsize=(14,4))

ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)
ax5 = fig.add_subplot(2,3,5)
ax6 = fig.add_subplot(2,3,6)

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))
ax4.plot(x, np.exp(x))
ax5.plot(x, np.log(x))
ax6.plot(x, np.tanh(x))
```
