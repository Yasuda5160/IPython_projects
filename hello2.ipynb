```python
x = np.linspace(0, 2 * np.pi, 100)

fig, axs = plt.subplots(1, 2, figsize = (8, 4), facecolor = 'gray', subplot_kw ={'facecolor': 'aqua'})

axs[0].plot(x, np.sin(x))
axs[1].plot(x, np.cos(x))
```

```python
fig, axs = plt.subplots(1, 2, figsize = (8, 4), constrained_layout = True, facecolor = 'gray', subplot_kw={'facecolor': 'aqua'})
axs[0].plot(x, np.sin(x))
axs[0].set_ylabel('ylabel')
axs[0].set_xlabel('xlabel')
axs[1].plot(x, np.cos(x))
axs[1].set_ylabel('ylabel')
axs[1].set_xlabel('xlabel')
```

```python
fig, axs = plt.subplots(2, 2, figsize = (8, 6), sharex = True, sharey = True, constrained_layout = True)

axs[0, 0].plot(np.sin(x), np.cos(x))
axs[0, 1].plot(np.sin(x), np.cos(2 * x))
axs[1, 0].plot(np.sin(x), np.cos(3 * x))
axs[1, 1].plot(np.sin(x), np.cos(4 * x))
```

```python
fig, axs = plt.subplots(2, 2, figsize = (8, 6), sharex = True, sharey = True, constrained_layout = True)
for i, ax in enumerate(axs.flat, 1):
    ax.plot(np.sin(x), np.cos(i * x))
```

```python
fig = plt.figure(figsize = (8, 6), constrained_layout = True)

gs = fig.add_gridspec(2, 3, width_ratios = [1, 1, 2], height_ratios = [1, 3])

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])
ax4 = fig.add_subplot(gs[0, 2])

ax1.set_title('gs[0, 0]')
ax2.set_title('gs[0, 1]')
ax3.set_title('gs[1, :]')
ax4.set_title('gs[0, 2]')
```

```python
gs_kw = {'width_ratios': [1, 3], 'height_ratios': [1, 2]}
fig, axs = plt.subplots(2, 2, figsize = (8, 6), gridspec_kw= gs_kw, constrained_layout = True)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

y1 = y + 1
y2 = y + 2

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x, y, '-.', lw = 1.5, label = 'dashdot')
ax.plot(x, y1, '--', lw = 2.5, label = 'dashed')
ax.plot(x, y2, ':', lw = 4.0, label = 'dotted')

ax.legend()
```

```python
fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x, y, 'k--')
ax.plot(x, y1, c = (0, 0, 1))
ax.plot(x, y2, 'gray')
```

```python
fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x[::10], y[::10], 'o', ms = 10, mfc = 'w', mew = 2.0, mec = 'b')
ax.plot(x[::10], y1[::10], '*--', c = '0.5', ms = 15)
ax.plot(x[::10], y2[::10], 'k:', marker = r'$clubsuit$', ms = 15)
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y = np.exp(-x) * np.sin(x)

plt.rcParams['font.family'] = 'Yu Mincho'

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x, y)
ax.grid()

ax.annotate('Japanese font', xy = (3, 0.15), fontsize = 18, c = 'b')
```
