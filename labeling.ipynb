```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, constrained_layout = True)
axs[0].plot(x, y1, '--', c = (1, 0, 0))
axs[1].plot(x, y2, '-.', c = (0, 1, 0))

fig.suptitle('Results of Our Calculations')

axs[0].set_title('No.1')
axs[1].set_title('No.2')
axs[0].set_xlabel('t [s]')
axs[1].set_xlabel('t [s]')
axs[0].set_ylabel('x [m]')
axs[1].set_ylabel('x [m]')
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sinc(x)
y2 = np.sinc(2 * x)

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x, y1, 'v--', ms = 4, label = 'sinc(x)')
ax.plot(x, y2, '*--',ms = 5, label = 'sinc(2x)')

ax.legend(fontsize = 12)
```

```python
coords = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(coords, coords)
Z = np.sin(X) * np.cos(Y)

fig, ax = plt.subplots(constrained_layout = True, subplot_kw= dict(aspect = 'equal'))

cs = ax.contourf(X, Y, Z, cmap = 'cividis')
cb = fig.colorbar(cs)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_label('z')
```

```python
from mpl_toolkits.mplot3d import Axes3D

coords = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(coords, coords)
Z = np.exp(- (X**2 + Y**2))

fig, axs = plt.subplots(1, 2, figsize = (8, 4), constrained_layout = True, subplot_kw= {'projection': '3d'})

surf = axs[0].plot_surface(X, Y, Z, rstride = 5, cstride = 5, cmap = 'civids')
fig.colorbar(surf, ax = axs[0], shrink = 0.6)

norm = plt.Normalize(Z.min(), Z.max())
colors = plt.cm.cividis(norm(Z))

surf = axs[1].plot_surface(X, Y, Z, rstride = 5, cstride = 5, facecolors = colors, shade = False)

surf.set_facecolor((0, 0, 0, 0))

axs[1].set_proj_type('ortho')
```
